```python
## This script is to calculate PWV grid domain for the purpose of sensitivity analyses
## ;;; move manually to stratus
## Created by C. Bayu Risanto, S.J. (23 February 2024)
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import random
import pandas as pd
import xarray as xr
from scipy.interpolate import griddata
import gzip
from netCDF4 import Dataset,num2date
import time
import glob
from get_PWVgrid import get_griddedPWV_abs
```

```python
## read data
dirin = '20210815_06Z_40mem_06Z_0035_6cyc_noassim'
for iens in range(40):   #set to 40
    post = 'wrfout_d02_2021-08-15_*'
    path = '/net/stratus/st8/bayu/ARIZONA-WRF/DART_runs/'
    wrffiles = sorted(glob.glob(path+dirin+'/'+'mem'+str(iens+1).zfill(2)+'/'+post))
    dir_out = '/home/bayu/Documents/ARIZONA-WRF/wrf_out/'+dirin+'/'
    
    ## read hourly wrfout
    for ih in range(len(wrffiles)):
        data = xr.open_dataset(wrffiles[ih])
        time = data.Times.values
        timee = time.astype('U19')
        pwv,xlon,xlat = get_griddedPWV_abs(wrffiles[ih])
        
        # create the nc file
        date = wrffiles[ih][-19:-3]
        dir_out = '/home/bayu/Documents/ARIZONA-WRF/wrf_out/'+dirin+'/'+'mem'+str(iens+1).zfill(2)+'/'
        fname = dir_out + 'PWV_d02_'+ date + '.nc'
        dat = Dataset(fname,'w',format='NETCDF3_CLASSIC')

        # add dimensions
        Time = dat.createDimension('Time',None)
        south_north = dat.createDimension('south_north',405)
        west_east = dat.createDimension('west_east',684)
        
        # add variables
        XTIME = dat.createVariable('XTIME','S1',('Time',))
        XLAT = dat.createVariable('XLAT','f4',('south_north','west_east',))
        XLONG = dat.createVariable('XLONG','f4',('south_north','west_east'))
        PWV = dat.createVariable('PWV','f4',('south_north','west_east',))
        
        # add units and variable attributes
        PWV.units = 'mm'
        PWV.coordinates = 'XLONG XLAT XTIME'
        PWV.description = 'grid scale precipitable water vapor'

        XTIME.units = 'hours'
        XTIME.calendar = 'gregorian'
        XTIME.description = date

        XLAT.units = 'degree_north'
        XLONG.units = 'degree_east'
        XLAT.long_name = 'latitude'
        XLONG.long_name = 'longitude'
        XLAT.axis = 'Y'
        XLONG.axis = 'X'
        XLAT.coordinates = 'XLONG XLAT'
        XLONG.coordinates = 'XLONG XLAT'
        
        # assign latitude and longitude
        XLAT[:,:] = xlat
        XLONG[:,:] = xlon

        # assign data values
        PWV[:,:] = pwv

        print('var size after adding first data',PWV.shape)
        print('save ',fname)

        dat.close() 
```
