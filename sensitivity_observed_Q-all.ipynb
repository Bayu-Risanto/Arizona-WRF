```python
## this script is to plot the beta (sensitivity of obs Qvap to obs PWV)
## data is processed in sensitivity_observed_Q-TUS.ipynb
## created by C. Bayu Risanto, S.J. (21 October 2025)
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
from numpy.linalg import inv
from scipy.io import loadmat
import xarray as xr
from datetime import date, datetime, timedelta
from gengc import gencpr
from calc_gc import calc_gc, calc_gc_bayu
import matplotlib.pyplot as plt
from scipy.io import loadmat,savemat 
```

```python
## read data
link = '/secchi/bayu/DATA/Gaspari-Cohn_project/sensitivity_beta/'
d1 = loadmat(link + 'sens_obsQvap_to_obsPWV_TUS_pro2021.mat')
b1 = np.squeeze(d1['beta']); m1 = np.squeeze(d1['mz'])
d2 = loadmat(link + 'sens_obsQvap_to_obsPWV_FGZ_pro2021.mat')
b2 = np.squeeze(d2['beta']); m2 = np.squeeze(d2['mz'])
d3 = loadmat(link + 'sens_obsQvap_to_obsPWV_EPZ_pro2021.mat')
b3 = np.squeeze(d3['beta']); m3 = np.squeeze(d3['mz'])
d4 = loadmat(link + 'sens_obsQvap_to_obsPWV_ABQ_pro2021.mat')
b4 = np.squeeze(d4['beta']); m4 = np.squeeze(d4['mz'])
```

```python
## average
bmean = (b1 + b2 + b3 + b4) / 4
mmean = (m1 + m2 + m3 + m4) / 4
```

```python
## Plot here ###
fig,ax1 = plt.subplots(1, 1, figsize=(5, 8))
ax1.plot(b4*1000,m4, marker='o',label='ABQ')
ax1.plot(b3*1000,m3, marker='o',label='EPZ')
ax1.plot(b2*1000,m2, marker='o',label='FGZ')
ax1.plot(b1*1000,m1, marker='o',label='TUS')
ax1.plot(bmean*1000,mmean,'--k',label='average')
ax1.annotate(r'$\beta$',(0.15,11.),fontsize=25, fontname='serif',
            bbox=dict(boxstyle='square',fc='w',ec='w'))
ax1.legend(loc=1)
ax1.set_ylim(0,12.5)
ax1.set_xlim(0,0.35)
ax1.set_title('Sens obs Qvap (mixrat) to obs PWV 2021')
    
ax1.set_xlabel('(g H2O/kg dry air) per mm')
ax1.set_ylabel('height km MSL')
ax1.grid(color='black', linestyle='dotted')

## SAVE
#dir_out = '/secchi/bayu/PLOTS/Gen-GASPARI-COHN_PRO/OBS/'
#plotfile  = 'Beta_QVAP2PWV_abq_epz_fgz_tus_12U_789_2021.png' ## Change hour accordingly!!!!!!!!
#sf = fig.savefig(dir_out+plotfile, dpi=300, bbox_inches='tight')

plt.show()
plt.close()
```
