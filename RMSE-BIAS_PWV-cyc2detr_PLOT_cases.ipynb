```python
## get mean RMSE across cases
da_x = np.arange(0,len(Fct_a),2) 
mrmse_da = np.mean(rmse_a[da_x],axis=0) 
no_x = np.arange(1,len(Fct_a),2) 
mrmse_no = np.mean(rmse_a[no_x],axis=0)
```

```python
## Get the pwv bias
isite,itime = Fct_exp01.shape
dif_a = [[Fct_a[j,:,i] - Obs_pwv_06[j,:,i] for i in range(itime)] for j in range(len(Fct_a[:,0,0]))]
dif_a = np.asarray(dif_a)
mu_dif_a = np.nanmean(dif_a,axis=2)

## get the mean across cases
da_x = np.arange(0,len(mu_dif_a),2) 
mdif_da = np.mean(mu_dif_a[da_x],axis=0) 
no_x = np.arange(1,len(mu_dif_a),2) 
mdif_no = np.mean(mu_dif_a[no_x],axis=0)
```
